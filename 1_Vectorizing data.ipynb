```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tqdm import tqdm
```

```python
categories=['Bathroom','Living room']
files=['file1','file2','file3','file4']
datadir="dataset"
```

```python
IMG_SIZE=150
X = []
y=[]
```

```python
for category in categories:  # do dogs and cats
    path = os.path.join(datadir,category)  # create path to dogs and cats
    for file in files:  # iterate over each image per dogs and cats
        path2=os.path.join(path,file)
        print(path2)
        for img in os.listdir(path2):
            path3=os.path.join(path2,img)
            try:
                img_array = cv2.imread(path3)  # convert to array
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                X.append(new_array)
                if category=='Bathroom':
                    y.append([0])
                else:
                    y.append([1])
            except Exception as e:  # in the interest in keeping the output clean...
                pass
```

```python
X=np.array(X)
y=np.array(y)
```

```python
classes=['Bathroom','Living room']
def plot_sample(X, y, index):
    plt.figure(figsize = (20,3))
    plt.imshow(X[index])
    plt.xlabel(classes[y[index][0]])
    plt.show()
```

```python
import random
for i in range(0,10):
    plot_sample(X, y, random.randrange(0,4013))
    plot_sample(X, y, random.randrange(4014,8187))
```
